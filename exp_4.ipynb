```python
import random
import pandas as pd
from mlxtend.frequent_patterns import apriori , association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
all_items = ['milk', 'bread', 'butter', 'jam', 'eggs', 'cheese']
```

```python
random.seed(50)
dataset = []
for _ in range(10):
    transaction_size = random.randint(2, 4)
    transaction = random.sample(all_items, transaction_size)
    dataset.append(transaction)
```

```python
def encode_transactions(dataset, items):
    encoded_data = []
    for transaction in dataset:
        encoded_row = {item: (item in transaction) for item in items}
        encoded_data.append(encoded_row)
    return pd.DataFrame(encoded_data)
```

```python
df = encode_transactions(dataset, all_items)
```

```python
frequent_itemsets = apriori(df,min_support=0.5,use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets,metric='lift',min_threshold=1.0)
```

```python
print(dataset)
print('\n')
print("\n Frequent Itemsets \n")
print(frequent_itemsets)
print('\n')
print("\n Association Rules \n")
print(rules[['antecedents', 'consequents', 'support', 'lift']])
```

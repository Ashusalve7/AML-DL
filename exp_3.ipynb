```python
import random
import pandas as pd
from mlxtend.frequent_patterns import apriori , association_rules
```

```python
def random_data(n):
    return [random.rand(True,False)for _ in range(n)]
```

```python
data = {
    'Milk': random_data(10),
    'Bread': random_data(10),
    'Eggs': random_data(10),
    'Butter': random_data(10),
    'Yogurt': random_data(10),
    'Cheese': random_data(10),
    'Apples': random_data(10),
    'Bananas': random_data(10),
    'Potatoes': random_data(10)
}
```

```python
df = pd.DataFrame(data)
```

```python
df.head(10)
```

```python
frequent_itemsets = apriori(df,min_support=0.5,use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets,metric='confidence',min_threshold=0.6)
```

```python
print(rules[['antecedents', 'consequents', 'support', 'lift']])
```
